# Matplotlib

Matplotlib adalah sebuah library Python yang digunakan untuk membuat visualisasi data dalam bentuk grafik atau plot. Library ini sangat populer di kalangan ilmuwan data, peneliti, dan pengembang software karena kemampuannya yang kuat untuk menghasilkan visualisasi yang menarik dan informatif. Dengan menggunakan Matplotlib, pengguna dapat membuat berbagai jenis grafik seperti grafik garis, grafik batang, grafik scatter, grafik histogram, dan banyak lagi.

Matplotlib menyediakan antarmuka yang fleksibel dan dapat disesuaikan, sehingga pengguna dapat mengatur berbagai elemen dalam plot seperti sumbu, label, judul, warna, dan gaya garis. Library ini juga mendukung fitur seperti penggabungan beberapa plot dalam satu gambar, anotasi, tampilan 3D, dan menyimpan plot dalam berbagai format file, seperti PNG, PDF, atau SVG.

Matplotlib bekerja dengan baik dengan banyak library dan framework lain dalam ekosistem Python, seperti NumPy dan Pandas. Selain itu, terdapat juga paket-paket lain yang dibangun di atas Matplotlib, seperti Seaborn, yang menyediakan fungsionalitas tambahan dan tampilan yang lebih cantik untuk visualisasi data.

Dengan menggunakan Matplotlib, pengguna dapat dengan mudah menghasilkan visualisasi yang berkualitas tinggi untuk menjelajahi dan memahami data mereka, serta mempresentasikan temuan mereka kepada orang lain.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y)
```

# API for plot 

```python
plt.plot(x,y,color='green',linewidth=5, linestyle='dotted')
```
